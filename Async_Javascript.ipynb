##1. Demonstrate JavaScript's Single-Threaded Nature
Question:
  Write an example to show that JavaScript is single-threaded by creating two competing tasks, one that blocks the event loop and another async function that waits for a promise.
- Javascript is single-threaded which can be evaluated by given code example. First it register a promise with takes 5 seconds to resolve. Then it starts a massive for loops which makes the tread busy for approx 10 seconds. So the promise resolve handler waits till the for loop finishes it work. It proves that javascript is single threaded, have only onle call stack and only one event loop.

  ```
  function eventLoopBlocker() {
    for (let i = 0; i < 10000000000; i++) {}
  }

  function myPromise() {
    return new Promise(function (resolve, reject) {
      console.log("Promise started");

      setTimeout(() => {
        resolve("Some value");
      }, 5000);
    });
  }

  // Test single thread nature
  console.log("Code start");
  console.log("Create promise and register a resolve callback");
  myPromise().then(function promiseHandler(value) {
    console.log("Promise resolved: ", value);
  });

  console.log("Call a massive for loop which blocks the thread until finishes");
  eventLoopBlocker();
  console.log("Event loop blocker completed!");

  console.log("Code Ended");
  ```

Runtime get the data from the server through AJAX. Then give it to javascript to finish it work.

## 3. Chaining Promises with setTimeout
 Modify the delay function to chain multiple promises so that three messages are logged in sequence with delays.

  ```
  console.log("Code start....");
  ```

## 4. What are the different states of a Promise, and how do they transition.
- There are 3 different states in promised. i. Pending ii. Resolved iii. Rejected.
  * i. Pending: This is the running state of promise when promise doing its job like getting data from server or something else. This is the default state of promise when its get created.
  * ii. Resolved: This is the state when a promise completes its job without any errors. we have to call a function resolve to make the promise to go in this state otherwise it will be in Pending state forever.
  * iii. Rejected: This is the state when a promise is finised with some error. We can make a promise to come in this state by calling reject function.

  Example:
  ```
  function myPromise() {
    return new Promise(function promiseResolver1(resolve, reject) {
      setTimeout(() => {
        const number = Math.floor(Math.random() * 101);
        if (number % 2 === 0) {
          resolve(number);
        } else {
          reject(number);
        }
      }, 2000);
    });
  }

  console.log("Code start...");
  const promise = myPromise();
  console.log("Promise created");
  console.log("Current Promise state: ", promise);

  promise.then(
    function resolveHandler(value) {
      console.log("Promise resolved successfully with data:", value);
      console.log("Current Promise state: ", promise);
    },
    function rejectHandler(value) {
      console.log("Promise rejected with data:", value);
      console.log("Current Promise state: ", promise);
    }
  );
  ```

## 5. How does the JavaScript event loop handle Promises differently from setTimeout?
- The JavaScript event loop treats Promises and setTimeout differently by placing them in different queues: Promises (and queueMicrotask) are added to the microtask queue. setTimeout, setInterval, setImmediate, etc. are placed in the macrotask queue (also called the task queue). So, if both setTimeout and promise handler is waiting, Promise handler will execute first then settimeout callback. It can be proved by given code example:
  ```
  function startTimer() {
    setTimeout(() => {
      console.log("Timer Finished!");
    }, 0);
    console.log("Timer Started");
  }

  function startPromise() {
    new Promise((resolve, reject) => {
      resolve("Successful!");
    }).then((value) => {
      console.log("Promise resolved callback received with data: ", value);
    });
  }

  console.log("Code start....");
  startTimer();
  startPromise();
  console.log("Code ends");
  ```
